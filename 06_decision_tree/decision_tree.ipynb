```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('salaries.csv')
df.head()
```

```python
x=df.iloc[:,0:3]
x.head(3)
```

```python
y=df['salary_more_then_100k']
y.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x=x.apply(le.fit_transform)
```

```python
x.head(3)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
model.predict([[1,1,0]])
```

###### Gini is the impurity in the Data and A gini score greater than zero implies that samples contained within that node belong to different classes. A gini score of zero means that the node is pure, that within that node only a single class of samples exist. 

```python
fig,ax=plt.subplots(figsize=(30,30))
tree.plot_tree(model.fit(x,y),filled=True)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(20,20))
tree.plot_tree(model.fit(x_train,y_train),filled=True)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(7,7))
tree.plot_tree(model.fit(x_test,y_test),filled=True)
plt.show()
```
